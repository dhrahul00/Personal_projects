```python
from neural_network import Network,FC_layer,AC_layer,Activation
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate synthetic dataset
np.random.seed(101)
X = np.linspace(-5, 5, 100).reshape(-1, 1)
y = (np.sin(X) + np.random.normal(0, 0.1, X.shape)).reshape(-1,1) # to consitent with the model

# Define the neural network
network = Network()
network.add(FC_layer(1, 10))
network.add(AC_layer(Activation().tanh, Activation().tanh_prime))
network.add(FC_layer(10, 1))

# Train the network
network.fit(X, y, epoch=500, learning_rate=0.01, batch_size=32, optimizer="ADAM", loss_type="MSE", regularization=None)

# Make predictions on the training data
predictions = network.predict(X)

# Plot the original data and predictions
plt.scatter(X, y, label="Original Data")
plt.plot(X, predictions, color='red', label="Predictions")
plt.xlabel("X")
plt.ylabel("y")
plt.title("Curve Fitting")
plt.legend()
plt.show()

# Plot the error curve
plt.plot(network.error_list)
plt.xlabel("Epoch")
plt.ylabel("Error")
plt.title("Training Error")
plt.show()
```
